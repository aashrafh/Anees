```python
import sys
import time
import pandas as pd
from gensim.corpora import WikiCorpus
```

```python
def make_corpus(in_f, out_f):
	"""Convert Wikipedia xml dump file to text corpus"""
	output = open(out_f, 'w')
	wiki = WikiCorpus(in_f)

	i = 0
	for text in wiki.get_texts():
		output.write(bytes(' '.join(text), 'utf-8').decode('utf-8') + '\n')
		i = i + 1
		if (i % 10000 == 0):
			print('Processed ' + str(i) + ' articles')
	output.close()
	print('Processing complete!')
```

```python
input      = "arzwiki-20211220-pages-articles-multistream.xml.bz2"
outupt = "/content/arzwiki.txt"
make_corpus(input, outupt)
```

```python
def load_corpus(input_file):

    """Loads corpus from text file"""

    print('Loading corpus...')
    time1 = time.time()
    corpus = input_file.read()
    time2 = time.time()
    total_time = time2-time1
    print('It took %0.3f seconds to load corpus' %total_time)
    return corpus
```

```python
input_file = open("/content/arzwiki.txt", 'r')
corpus = load_corpus(input_file)
```
