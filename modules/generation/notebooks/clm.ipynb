```python
os.chdir("/content")
from sklearn.model_selection import train_test_split
from arabert.preprocess import ArabertPreprocessor

MODEL_NAME = "aubmindlab/aragpt2-base"
arabert_prep = ArabertPreprocessor(model_name=MODEL_NAME)

with open('hageen_corpus.txt', 'r') as data:
  dataset = ["<|bot|>" + arabert_prep.preprocess(x.strip()) for x in data.readlines()]
```

```python
train, eval = train_test_split(dataset, train_size=.9, test_size=0.1, random_state=2022)
print("training size:" + str(len(train)))
print("Evaluation size: " + str(len(eval)))

with open('train_tmp.txt', 'w') as file_handle:
  file_handle.write("<|eot|>".join(train))

with open('eval_tmp.txt', 'w') as file_handle:
  file_handle.write("<|eot|>".join(eval))
```

```python
%%bash
python /content/transformers/examples/pytorch/language-modeling/run_clm.py \
    --model_name_or_path "aubmindlab/aragpt2-base" \
    --train_file "train_tmp.txt" \
    --validation_file "eval_tmp.txt" \
    --per_device_train_batch_size 1 \
    --per_device_eval_batch_size 1 \
    --do_train \
    --do_eval \
    --num_train_epochs 10 \
    --output_dir "/content/drive/MyDrive/run_clm_hageen" \
    --overwrite_output_dir
```

```python
from transformers import GPT2TokenizerFast, pipeline
from transformers import GPT2LMHeadModel


model = GPT2LMHeadModel.from_pretrained("/content/drive/MyDrive/run_clm_hageen")
tokenizer = GPT2TokenizerFast.from_pretrained("aubmindlab/aragpt2-base")

text = "ازيك"
text_clean = arabert_prep.preprocess(text)

generation_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer)

generation_pipeline(text,
    pad_token_id=tokenizer.eos_token_id,
    num_beams=10,
    max_length=200,
    top_p=0.9,
    repetition_penalty = 3.0,
    no_repeat_ngram_size = 3)[0]['generated_text']
```
